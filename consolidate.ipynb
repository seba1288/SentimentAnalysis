```python
import os
import pandas as pd
import glob
```

```python
files = [f'./Opinions/{file}' for file in os.listdir('./Opinions/') if file.endswith('.csv')]
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df.to_csv('./Opinions/opinions_full.csv', index=False)
```
